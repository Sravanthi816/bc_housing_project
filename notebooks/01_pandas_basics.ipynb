```python
df["median_housing_price"].mean()
df["median_housing_price"].max()
df["median_housing_price"].min()
df["median_housing_price"].sum()
df["median_housing_price"].count()
df["median_housing_price"].std()
df["median_housing_price"].var()
df.groupby("cma_name")["median_housing_price"].mean()
df.groupby("cma_name")["median_housing_price"].max()
df.groupby("cma_name")["median_housing_price"].min()
df.groupby("cma_name")["median_housing_price"].sum()
df.groupby("cma_name")["median_housing_price"].count()
df.groupby("cma_name")["median_housing_price"].agg(["mean", "max", "min", "sum", "count", "std", "var"])
```

```python
grouped_df = df.groupby("cma_name")["median_housing_price"].mean()
grouped_df.reset_index()
df.groupby("cma_name")["median_housing_price"].mean()\
.sort_values(ascending=False)\
.head(5).reset_index()
grouped_df = df.groupby("cma_name")["median_housing_price"].mean().reset_index()
result=grouped_df[grouped_df["median_housing_price"]>100000].head(10)
print(result)


grouped_df = df.groupby("cma_name")["mortgage_payment_percent_income"].mean()
result = grouped_df[grouped_df > 35].head(10).reset_index()
result


```

```python
df.groupby(["cma_name", "housing_type"])["median_housing_price"].mean().head(10).reset_index()

df.pivot_table(
    index="cma_name",
    columns="housing_type",
    values="median_housing_price",
    aggfunc="mean"
).isnull().sum()
df.pivot_table(
    index="cma_name",
    columns="housing_type",
    values="median_housing_price",
    aggfunc="sum"
).isnull().sum()
df.pivot_table(
    index="cma_name",
    columns="housing_type",
    values="median_housing_price",
    aggfunc="max"
).fillna(0)
```

```python
avg_price_df =(df.groupby("cma_name")["median_housing_price"]
               .mean()
               .sort_values(ascending=True).
               head(10)
               )
avg_price_df.plot(kind="bar", 
                  figsize=(12,6),
                  xlabel="CMA Region", 
                  ylabel="Average Median Housing Price",
                  rot=45,grid=True,
                  color="blue",
                  title="Average Median Housing Price by CMA")
df.plot(kind="scatter", 
        x="median_housing_price", 
        y="mortgage_payment_percent_income",
        figsize=(12,6),
        rot=45,grid=True,
        color="blue",
        title="Average Median Housing Price by CMA")
vancouver_df = df[df["cma_name"] =="Vancouver"]
vancouver_trend = (
    vancouver_df.groupby("year")["median_housing_price"].mean()
)
vancouver_trend.plot(kind="line",
                     figsize=(12,6),
                     xlabel="Year",
                     ylabel="Average Median Housing Price",
                     rot=45,grid=True,
                     color="blue",
                     title="Average Median Housing Price Trend in Vancouver")   
```

```python
compare_df = df[(df["cma_name"]=="Vancouver") | 
                (df["cma_name"]=="Victoria")]

trend_df = compare_df.pivot_table(
    index="year",
    columns="cma_name",
    values="median_housing_price",
    aggfunc="mean"
)
trend_df.plot(kind="line",
            figsize=(12,6),
            xlabel="Year",
            ylabel="Average Median Housing Price",    
            rot=45,grid=True,
            color=["blue", "orange"],
            title="Vancouver vs Victoria Housing Trends")

```

```python
import matplotlib.pyplot as plt
vancouver_df = df[df["cma_name"]=="Vancouver"]
Vancouver_trend = ( vancouver_df.groupby("year")["median_housing_price"]
                   .mean() )
highest_year = vancouver_trend.idxmax()
highest_price = vancouver_trend.max()
plt.figure(figsize=(12,6))
plt.plot(Vancouver_trend.index, Vancouver_trend.values, color='blue')
plt.title("Average Median Housing Price Trend in Vancouver")
plt.xlabel("Year")
plt.ylabel("Average Median Housing Price")
plt.grid()
plt.annotate("highest price",(highest_year, highest_price), 
             xytext=(5, 10), textcoords='offset points',ha='center')
plt.show()
```

```python
import pandas as pd
```

```python

regions_df = pd.read_csv("../data/regions.csv")

interest_df = pd.read_csv("../data/yearly_interest_rates.csv")
regions_df.head()
interest_df.head()
```

```python
regions_df.columns
interest_df.columns
df.columns
```

```python
merge_df = pd.merge(df, regions_df, on = "cma_name" , how = "inner")
merge_df.head(10)
merge_df.columns
```

```python
final_df= pd.merge(df,interest_df, on="year", how="inner")
final_df.head()
final_df.columns
final_df.shape
```

```python
ultimate_df =pd.merge(merge_df,interest_df, on="year", how="inner")
ultimate_df.head()
ultimate_df.columns
ultimate_df.shape
```

```python
ultimate_df[[
    "median_housing_price",
    "average_interest_rate",
    "inflation_rate",
    "population",
    "unemployment_rate",
    "mortgage_payment_percent_income"
]].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
correlation_matrix = ultimate_df[[
    "median_housing_price",
    "average_interest_rate",
    "inflation_rate",
    "population",
    "unemployment_rate",
    "mortgage_payment_percent_income"
]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Housing Price and Economic Indicators")
plt.show()
```
